```python
import numpy as np
import pandas as pd


import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#To find the sales trend over the time
x= [1,2,3,4,5,6,7] #days of the week
y= [160,150,140,145,175,165,180] #sales of a company
plt.plot(x,y,'ro--')
plt.title('Sales of the week')
plt.xlabel('days of the week')
plt.ylabel('sales')
```
